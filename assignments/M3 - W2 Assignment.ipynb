Custom Age Grouping and Fare Analysis:

Create a function categorize_age_and_analyze_fare(df: pd.DataFrame) -> pd.DataFrame that:
Groups the Titanic dataframe by "Pclass".
Creates a new column "Age_Range" categorizing ages into custom bins (e.g., "Child", "Teen", "Adult", "Senior") based on arbitrary criteria you define beyond simple 5-year intervals.
Generates a summary table showing the count of passengers, average, and median fare for each "Age_Range" within each "Pclass".
Returns this summary table

```python
def categorize_age_and_analyze_fare(df: pd.DataFrame) -> pd.DataFrame:
    """
    Categorizes ages into custom bins and analyzes fare statistics within each Pclass.

    Parameters:
    - df (pd.DataFrame): Titanic DataFrame.

    Returns:
    - pd.DataFrame: Summary table with Pclass, Age_Range, Passenger_Count, Average_Fare, and Median_Fare.
    """

    # Define custom age bins and labels
    age_bins = [0, 12, 18, 35, 60, df['Age'].max()]
    age_labels = ["Child", "Teen", "Adult", "Senior", "Elderly"]

    # Create a new column "Age_Range" based on custom bins
    df['Age_Range'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

    # Group by "Pclass" and "Age_Range" and calculate count, average, and median fare
    summary_table = df.groupby(['Pclass', 'Age_Range']).agg({
        'PassengerId': 'count',
        'Fare': ['mean', 'median']
    }).reset_index()

    # Rename columns for better readability
    summary_table.columns = ['Pclass', 'Age_Range', 'Passenger_Count', 'Average_Fare', 'Median_Fare']

    return summary_table

```

Complex Family Size Analysis:

Create a function analyze_family_size_survival(df: pd.DataFrame) -> pd.DataFrame that:
Calculates "FamilySize" as the sum of "SibSp" and "Parch" plus one (for the passenger themselves).
Categorizes "FamilySize" into custom groups (e.g., "Solo", "Small", "Medium", "Large") with your own defined thresholds.
Creates a pivot table showing the survival rate and average fare for each "FamilySize" category, broken down by "Pclass".
Returns the pivot table

```python
def analyze_family_size_survival(df: pd.DataFrame) -> pd.DataFrame:
    """
    Analyzes survival rate and average fare based on custom family size categories and Pclass.

    Parameters:
    - df (pd.DataFrame): Titanic DataFrame.

    Returns:
    - pd.DataFrame: Pivot table with FamilySize, Pclass, Survival Rate, and Average Fare.
    """

    # Calculate FamilySize as the sum of SibSp and Parch plus one
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

    # Define custom family size bins and labels
    size_bins = [0, 1, 3, 5, df['FamilySize'].max()]
    size_labels = ["Solo", "Small", "Medium", "Large"]

    # Create a new column "FamilySize_Category" based on custom bins
    df['FamilySize_Category'] = pd.cut(df['FamilySize'], bins=size_bins, labels=size_labels, right=False)

    # Create a pivot table showing survival rate and average fare for each FamilySize category and Pclass
    pivot_table = df.pivot_table(index='FamilySize_Category', columns='Pclass', values=['Survived', 'Fare'], aggfunc={'Survived': 'mean', 'Fare': 'mean'})

    return pivot_table

```

```python
analyze_family_size_survival(titanic_df)
```

Survival rates are relatively high across family size categories.
"Large" family size in Pclass 1 has a survival rate lower than "Small" and "Medium."
The average fare is highest for "Large" family sizes, indicating potential economic differences.

Pclass 2:

Survival rates are generally high, with "Large" family size having a perfect survival rate (1.0) in Pclass 2.
"Large" family size in Pclass 2 has the highest average fare, suggesting a correlation between family size and economic status.

Pclass 3:

Survival rates are notably lower for larger family sizes, especially in the "Large" category.
The average fare is the highest for "Large" family size in Pclass 3, reflecting potential economic disparities.

Creates a pivot table showing the average and median age of passengers grouped by 'Sex', 'Pclass', and 'Embarked'.
The function should highlight any interesting patterns or anomalies in the data (e.g., unusually high or low ages for certain groups).

```python
def pivot_age_by_class_and_embarkation(df: pd.DataFrame) -> pd.DataFrame:
    """
    Creates a pivot table showing the average and median age of passengers
    grouped by 'Sex', 'Pclass', and 'Embarked'.

    Parameters:
    - df (pd.DataFrame): Titanic DataFrame.

    Returns:
    - pd.DataFrame: Pivot table with 'Sex', 'Pclass', 'Embarked', Average_Age, and Median_Age.
    """

    # Create a pivot table with average and median age
    age_pivot = df.pivot_table(index=['Sex', 'Pclass', 'Embarked'],
                               values='Age',
                               aggfunc=['mean', 'median']).reset_index()

    # Rename columns for better readability
    age_pivot.columns = ['Sex', 'Pclass', 'Embarked', 'Average_Age', 'Median_Age']

    # Highlight any interesting patterns or anomalies
    interesting_patterns = age_pivot[(age_pivot['Average_Age'] > 60) | (age_pivot['Median_Age'] < 10)]

    if not interesting_patterns.empty:
        print("Interesting Patterns or Anomalies:")
        print(interesting_patterns)
    else:
        print("No interesting patterns or anomalies found.")

    return age_pivot


```

Create a function clean_and_analyze_embarkation(df: pd.DataFrame) -> pd.DataFrame that:
Removes "Cabin" and "Ticket" columns.
Fills missing "Embarked" values with the most common (mode) port of embarkation.
Analyzes the survival rate and average "Fare" for passengers embarking from each port, including a count of passengers.
Returns a dataframe with this analysis.

```python
def clean_and_analyze_embarkation(df: pd.DataFrame) -> pd.DataFrame:
    """
    Cleans the DataFrame by removing 'Cabin' and 'Ticket' columns,
    fills missing 'Embarked' values with the mode, and analyzes
    survival rate and average 'Fare' for passengers embarking from each port.

    Parameters:
    - df (pd.DataFrame): Titanic DataFrame.

    Returns:
    - pd.DataFrame: Analysis of survival rate, average 'Fare', and passenger count for each port of embarkation.
    """

    # Remove 'Cabin' and 'Ticket' columns
    df_cleaned = df.drop(['Cabin', 'Ticket'], axis=1)

    # Fill missing 'Embarked' values with the mode using df.loc
    most_common_embarked = df_cleaned['Embarked'].mode().values[0]
    df_cleaned.loc[:, 'Embarked'] = df_cleaned['Embarked'].fillna(most_common_embarked)

    # Analyze survival rate, average 'Fare', and passenger count for each port of embarkation
    embarkation_analysis = df_cleaned.groupby('Embarked').agg({
        'Survived': 'mean',
        'Fare': 'mean',
        'PassengerId': 'count'
    }).reset_index()

    # Rename columns for better readability
    embarkation_analysis.columns = ['Embarked', 'Survival_Rate', 'Average_Fare', 'Passenger_Count']

    return embarkation_analysis

```

```python
clean_and_analyze_embarkation(titanic_df)
```

The survival rates vary across different ports of embarkation, suggesting potential differences in passenger demographics or boarding locations.

Average fares also show variations, indicating potential economic disparities among passengers from different embarkation ports.

Southampton (S) had the highest passenger count, likely due to being a major embarkation port.

Create a function fare_range_analysis(df: pd.DataFrame) -> pd.DataFrame that:

Uses pd.cut or pd.qcut to create custom fare bins (e.g., based on quartiles or specific fare thresholds).

Groups the dataframe by these "FareRange" bins and "Sex", then calculates the mean, minimum, maximum, and standard deviation of "Age", as well as the survival rate for each group.
Returns a dataframe with this grouped analysis

```python
def fare_range_analysis(df: pd.DataFrame) -> pd.DataFrame:
    """
    Performs fare range analysis by creating custom fare bins and grouping
    the DataFrame by these bins and 'Sex'. Calculates mean, minimum, maximum,
    standard deviation of 'Age', and survival rate for each group.

    Parameters:
    - df (pd.DataFrame): Titanic DataFrame.

    Returns:
    - pd.DataFrame: Grouped analysis with 'FareRange', 'Sex', 'Mean_Age', 'Min_Age',
      'Max_Age', 'Age_Std', and 'Survival_Rate'.
    """

    # Create custom fare bins using pd.qcut (you can adjust bins as needed)
    fare_bins = pd.qcut(df['Fare'], q=[0, 0.25, 0.5, 0.75, 1.0], labels=['Q1', 'Q2', 'Q3', 'Q4'], duplicates='drop')
    
    # Add 'FareRange' column to the DataFrame
    df['FareRange'] = fare_bins

    # Group by 'FareRange' and 'Sex', then calculate the requested statistics
    fare_analysis = df.groupby(['FareRange', 'Sex'],observed=False).agg({
        'Age': ['mean', 'min', 'max', 'std'],
        'Survived': 'mean'
    }).reset_index()

    # Rename columns for better readability
    fare_analysis.columns = ['FareRange', 'Sex', 'Mean_Age', 'Min_Age', 'Max_Age', 'Age_Std', 'Survival_Rate']

    return fare_analysis
```

```python
fare_range_analysis(titanic_df)
```

Generally, females tend to have higher survival rates across all fare ranges.

Survival rates for both genders are higher in higher fare ranges (Q4) compared to lower fare ranges (Q1).

Mean age tends to increase with higher fare ranges for both genders.
