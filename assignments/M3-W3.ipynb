###  How much in total did every Sales Agent make

```python
%%sql
select c.supportrepid as agent_id,concat(e.firstname," ",e.lastname) as agent_name,sum(i.total) as total_sales from invoices as i
inner join customers as c on c.customerid=i.customerid
inner join employees as e on e.employeeid=c.supportrepid
group by agent_id
```

### What is the sales agent that made the most sales in 2012

```python
%%sql
select  c.supportrepid as agent_id,concat(e.firstname," ",e.lastname) as agent_name,sum(i.total) as total_sales, substr(invoicedate,1,4) as year
from invoices as i
inner join customers as c on c.customerid=i.customerid
inner join employees as e on e.employeeid=c.supportrepid
where year ='2012'
group by agent_id
order by total_sales desc
limit 1
```

### Which sales agent made the most in sales overall?

```python
%%sql
select c.supportrepid as agent_id,concat(e.firstname," ",e.lastname) as agent_name,sum(i.total) as total_sales from invoices as i
inner join customers as c on c.customerid=i.customerid
inner join employees as e on e.employeeid=c.supportrepid
group by agent_id
order by agent_id
limit 1
```

### What is the number of customers assigned to each sales agent

```python
%%sql
select  c.supportrepid as agent_id,
concat(e.firstname," ",e.lastname) as agent_name, 
COUNT(DISTINCT c.customerid) AS num_customers
from invoices as i 
inner join customers as c on c.customerid=i.customerid
inner join employees as e on e.employeeid=c.supportrepid
group by agent_id
```

### Show the most purchased track of 2010

```python
%%sql
select t.name ,count(*) as purchase_count ,substr(invoicedate,1,4) as year from tracks as t
inner join invoice_items as i on t.trackid=i.trackid
inner join invoices as iv on iv.invoiceid=i.invoiceid
where year ='2010'
group by year
```

### Show the top 5 most purchased tracks ever.

```python
%%sql
select t.name as track_name ,count(*) as purchase_count from tracks as t
inner join invoice_items as i on t.trackid=i.trackid
group by track_name
order by purchase_count desc
limit 5
```

### What are the top 3 best selling artists?

```python
%%sql
select a.artistid,a.name,count(*) as count from artists as a
inner join albums using (artistid)
inner join tracks using (albumid)
inner join invoice_items using (trackid)
group by a.artistid
order by count desc
limit 3
```

### What is the most purchased Media Type?

```python
%%sql
select mediatypeid,m.name,count(*) as count
from media_types as m
inner join tracks as t using (mediatypeid)
inner join invoice_items as i using (trackid)
inner join invoices as iv using (invoiceid)
group by mediatypeid
order by count desc
limit 5
```

### Show the number of tracks purchased in all invoices that contain more than one genre

```python
%%sql

select iv.invoiceid,COUNT(DISTINCT t.trackid) AS num_tracks_purchased,COUNT(DISTINCT t.genreid) AS num_genres
from invoices AS iv
inner join invoice_items AS ii ON iv.invoiceid = ii.invoiceid
inner join tracks AS t ON ii.trackid = t.trackid
group by iv.invoiceid
having COUNT(DISTINCT t.genreid) > 1
order by num_genres desc
```
