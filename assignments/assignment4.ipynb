### I.    Write a function (with the appropriate dockstrings and typings) that will take as input a list of strings andwill return a dictionary of iterables, where each iterable contains the location of the key in the original list, for example, if the function takes as input the list [1,2,3,2,3,1] it should return {1:[0,5], 2:[1,3], 3[2,4]}.  There will be bonus points if you can do it without the use of any for loops.

```python
from typing import List, Dict

def dict_indices(input_list: List[str]) -> Dict[str, List[int]]:
    """
    Create a dictionary of iterables where each iterable contains the
    location of the key in the original list.

    Parameters:
    - input_list (List[str]): The input list of strings.

    Returns:
    - Dict[str, List[int]]: A dictionary where keys are unique elements
      from the input list, and values are lists of indices corresponding
      to the location of the key in the input list.
    """
    # Create a dictionary with default values as lists
    indices_dict = {key: [] for key in set(input_list)}

    # Use enumerate and a list comprehension to populate the dictionary
    [indices_dict[key].append(index) for index, key in enumerate(input_list)]

    return indices_dict

# Example 

input_list = [1, 2, 3, 2, 3, 1]
result = dict_indices(input_list)
print(result)
```

### I.    Consider the same list found here. Write a function (with the appropriate typings and dockstrings) that returns a list of the unique elements in the original list. Extra points if you can do this without the use of the built-in set function.

```python
from typing import List

def unique_elements(input_list: List[str]) -> List[str]:
    """
    Return a list of unique elements from the original list.

    Parameters:
    - input_list (List[str]): The input list of strings.

    Returns:
    - List[str]: A list containing unique elements from the input list.
    """
    unique_list = []
    [unique_list.append(element) for element in input_list if element not in unique_list]
    return unique_list

# Example usage:
input_list = [1, 2, 3, 2, 3, 1]
result = unique_elements(input_list)
print(result)
```
